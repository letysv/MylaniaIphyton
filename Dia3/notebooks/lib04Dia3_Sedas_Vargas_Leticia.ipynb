```python
from datetime import timedelta

def totimestamp(dt, epoch=datetime(1970,1,1)):
    td = dt - epoch
    # return td.total_seconds()
    return (td.microseconds + (td.seconds + td.days * 24 * 3600) * 10**6) / 1e6 
```

```python
vFechaFloat = totimestamp(vFechaRegistrada)
```
